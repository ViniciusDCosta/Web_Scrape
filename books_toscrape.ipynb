# Criando df_raw em uma celula so

Nessa parte vamos juntar todos os passos acima em um único Data Frame

```python
#list of url's
list1 = ['classics_6', 'science-fiction_16', 'humor_30', 'business_35']

#empty dataframe
df_raw = pd.DataFrame()

# Parameters
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36' }

for i in range(len(list1)):
    
    #url
    url = 'http://books.toscrape.com/catalogue/category/books/' + list1[i]
    #print(url)
    
    # Request url
    page = requests.get( url, headers=headers)

    #Beautiful soup object
    soup = BeautifulSoup( page.text, 'html.parser' )

    #================= Product Data =================
    products = soup.find( 'ol', class_= 'row')

    #-----------price------------
    product_list = products.find_all( 'p', class_='price_color')
    product_price = [p.get_text() for p in product_list]

    #-----------availability----------
    product_list = products.find_all( 'p', class_='instock availability')
    product_availability = [list(filter(None, p.get_text().split('\n'))) for p in product_list]
    for j in range(len(product_availability)):
        product_availability[j]  = product_availability[j][1]
        product_availability[j]  = product_availability[j].split('  ')
        product_availability[j]  = product_availability[j].pop()

    #-----------name-----------------
    product_name = []
    for a in products.find_all('a', title=True):
        product_name.append( a['title'] )

    #-----------Stars-----------------

    #product_list = products.find_all( 'article', class_='product_pod')
    product_stars = []
    for a in products.find_all('p', class_=True):
        product_stars.append( a['class'] )
    product_stars = product_stars[::3]
    for j in range(len(product_stars)):
        product_stars[j]  = product_stars[j][1]

    #-----------Category------------------
    product_category = []
    product_list = soup.find( 'div', class_='page-header action')
    product_category_1 = [p.get_text() for p in product_list][1]
    for j in range(len(product_stars)):
        product_category.append(product_category_1)


    # creating the dataframes
    df1 = pd.DataFrame([product_name, product_category, product_price, product_availability, product_stars]).T
    df1.columns = ['product_name', 'product_category', 'product_price', 'product_availability', 'product_stars']
    
    #joining the dataframe in the final df
    df_raw = pd.concat([df_raw, df1]).reset_index(drop=True)

#-------Creating the scrapy time column
df_raw['scrapy_datetime'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

df_raw.head()
```

# Limpeza de dados e Resultado Final

```python
df_data = df_raw.copy()

#product name
df_data['product_name'] = df_data['product_name'].str.replace( ' ', '_' ).str.lower()

#product category
df_data['product_category'] = df_data['product_category'].str.replace( ' ', '_' ).str.lower() 

#product availability
df_data['product_availability'] = df_data['product_availability'].str.replace( ' ', '_' ).str.lower() 

# product price
for x in range(len(df_data)):
    df_data['product_price'][x] = re.findall( r'\d+\.?\d+', df_data['product_price'][x])[0]
df_data['product_price'] = df_data['product_price'].astype( float )

#scrapy datetime
df_data['scrapy_datetime'] = pd.to_datetime(df_data['scrapy_datetime'], format='%Y-%m-%d %H:%M:%S')

#product star
df_data['product_stars'] = df_data['product_stars'].apply(w2n.word_to_num)

df_data.head(20)
```
